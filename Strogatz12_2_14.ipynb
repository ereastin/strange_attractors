```python
def transform1(a, x, y):
    x = x
    y = 1 + y - a * abs(x)
    return x, y
```

```python
def transform2(b, x, y):
    x = b * x
    y = y
    return x, y
```

```python
def transform3(x, y):
    x = y
    y = x
    return x, y
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def rectangle0(n):
    xtb = np.linspace(-5, 5, n)
    ytb = np.full(n, 3)
    xlr = np.full(n, 5)
    ylr = np.linspace(-3, 3, n)
    return xtb, ytb, xlr, ylr
```

```python
def lozitransform(a, b, shape):
    k, l, m, n = shape
    plt.figure(1)
    plt.grid()
    plt.xlim((-10, 10))
    plt.ylim((-10, 10))
    plt.xlabel('x')
    plt.ylabel('y')
    plt.plot(k, l, 'k')
    plt.plot(k, -l, 'k')
    plt.plot(-m, n, 'k')
    plt.plot(m, n, 'k')
    # plt.savefig('T.png')
    k10, ltop1 = transform1(a, k, l)
    k11, lbottom1 = transform1(a, k, -l)
    mleft1, n10 = transform1(a, -m, n)
    mright1, n11 = transform1(a, m, n)
    plt.figure(2)
    plt.grid()
    plt.xlim((-10, 10))
    plt.ylim((-10, 10))
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title(r"$T': x'=x, \: y'=1+y-ax^2$")
    plt.plot(k10, ltop1, 'k')
    plt.plot(k11, lbottom1, 'k')
    plt.plot(mleft1, n10, 'k')
    plt.plot(mright1, n11, 'k')
    # plt.savefig('T1.png')
    k20, ltop2 = transform2(b, k10, ltop1)
    k21, lbottom2 = transform2(b, k11, lbottom1)
    mleft2, n20 = transform2(b, mleft1, n10)
    mright2, n21 = transform2(b, mright1, n11)
    plt.figure(3)
    plt.grid()
    plt.xlim((-10, 10))
    plt.ylim((-10, 10))
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title(r"$T'': x''=bx', \: y''=y'$")
    plt.plot(k20, ltop2, 'k')
    plt.plot(k21, lbottom2, 'k')
    plt.plot(mleft2, n20, 'k')
    plt.plot(mright2, n21, 'k')
    # plt.savefig('T2.png')
    plt.figure(4)
    plt.grid()
    plt.xlim((-10, 10))
    plt.ylim((-10, 10))
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title(r"$T''': x'''=y'', \: y'''=x''$")
    plt.plot(ltop2, k20, 'k')
    plt.plot(lbottom2, k21, 'k')
    plt.plot(n20, mleft2, 'k')
    plt.plot(n21, mright2, 'k')
    # plt.savefig('T3.png')
```

```python
lozitransform(1, 0.4, rectangle0(500))
plt.show()
```
