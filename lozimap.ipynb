```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def LoziMap(a, b, x0, y0, N):
    x, y = np.zeros(N), np.zeros(N)
    x[0], y[0] = x0, y0
    for i in range(1, N, 1):
        x[i] = 1 + y[i - 1] - a * abs(x[i - 1])
        y[i] = b * x[i - 1]
    return x, y
```

```python
def plotLoziMap(a, b):
    x, y = LoziMap(a, b, 0, 0, 10000)
    plt.plot(x[5:], y[5:], '.', color='k', markersize=.5)
    plt.xlabel('x')
    plt.ylabel('y')
    # plt.savefig('loziattractor.png')
    plt.show()
```

```python
plotLoziMap(1.7, 3 - 2 * np.sqrt(2))
```
