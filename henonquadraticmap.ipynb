```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.style
import matplotlib as mpl
mpl.style.use('classic')
```

```python
def HAPQMap(a, x0, y0, N): # Henon's Area-Preserving Quadratic Map
    x, y = np.zeros(N), np.zeros(N)
    x[0], y[0] = x0, y0
    for i in range(1, N, 1):
        x[i] = x[i - 1] * np.cos(a) - (y[i - 1] - x[i - 1] ** 2) * np.sin(a)
        y[i] = x[i - 1] * np.sin(a) + (y[i - 1] - x[i - 1] ** 2) * np.cos(a)
    return x, y
```

```python
def plotHAPQMap(value, x1, x2, y1, y2, N):
    x0 = np.linspace(-1, 1, 10) # mapping with initial conditions -1 <= x, y <= 1
    y0 = np.linspace(-1, 1, 10)
    coords = np.zeros((len(x0), len(y0)), dtype=tuple) # initializing array [S] of initial conditions
    alpha = np.arccos(value) # angle given in problem
    plt.figure()
    for m in range(0, len(x0)): # [S]_{mn} matrix value 
        for n in range(0, len(y0)):
            coords[m][n] = (x0[m], y0[n])
    for i in range(0, len(coords[0])): 
        for j in range(0, len(coords[1])):
            p, q = coords[i][j] # unpacking x, y values
            x, y = HAPQMap(alpha, p, q, 10000) # calculating x, y arrays over 5000 iterations
            plt.plot(x, y, '.', color='k', markersize=N)
    plt.xlim((x1, x2))
    plt.ylim((y1, y2))
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend(title=r'$\cos(\alpha) = $'+str(round(value, 3)))
    #plt.savefig('HAPQMapa24zoom.png')
    plt.show()
```

```python
plotHAPQMap(0.24, -1, 1, -1, 1, .03)
plotHAPQMap(0.22, -1, 1, -1, 1, .03)
plotHAPQMap(0.24, .4, .7, 0, .4, .3)

```
